```python
class CleanData:
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.check = True
        self.X_category = self.dataframe.select_dtypes(include=['object'])
        self.X_Numeric = self.dataframe.select_dtypes(include=['number'])

    def IsInconsistence(self):
        if self.dataframe.isna().sum().sum() == 0:
            self.check = False
            return self.dataframe
        else:
            pipe_num = Pipeline([
                ('pipe_num', SimpleImputer(strategy='median'))
            ])

            pipe_cat = Pipeline([
                ('pipe_cat', SimpleImputer(strategy='most_frequent'))
            ])

            merge_dataframe = pd.concat([self.X_Numeric, self.X_category], axis=1)

            pipe_full = ColumnTransformer([
                ('pipe_num', pipe_num, list(self.X_Numeric.columns)),
                ('pipe_cate', pipe_cat, list(self.X_category.columns))
            ])

            my_dataframe_transformer = pipe_full.fit_transform(merge_dataframe)
            my_full_dataframe_clean = pd.DataFrame(
                my_dataframe_transformer, columns=merge_dataframe.columns,
                index=merge_dataframe.index
            )

            self.dataframe=my_dataframe_transformer.copy()

            return self.dataframe
```

```python
class PreprocessData(CleanData):
  def __init__(self,dataframe):
    super().__init__(dataframe)
    self.std=StandardScaler()

  def Preprocess(self):
    self.dataframe[self.X_Numeric.columns[:-1]]=self.std.fit_transform(
        self.dataframe[self.X_Numeric.columns[:-1]]
    )

    return self.dataframe
```

```python
class Split_training_data:
  def __init__(self,dataframe,percentile):
    self.dataframe=dataframe
    self.percentile=percentile

  def Split(self,y):
    try:
      y=str(y)
      X=self.dataframe.drop(
          columns=[y],
          axis=1
      )
      y=self.dataframe[y]

      X_train,X_test,y_train,y_test=train_test_split(
          X,y,
          test_size=self.percentile,
          random_state=24
      )
      X_train_Numeric=X_train.select_dtypes(include=['number'])
      X_train_Category=X_train.select_dtypes(include=['object'])
      X_test_Numeric=X_test.select_dtypes(include=['number'])
      X_test_Category=X_test.select_dtypes(include=['object'])

      return X_train_Numeric,X_test_Numeric,X_train_Category,X_test_Category,y_train,y_test

    except (ValueError,TypeError) as e :
      raise type(e)(f'Error type')
```

```python
class CreateArquitectureModel(Split_training_data):
    def __init__(self, dataframe, percentile):
        self.out_of_vocabulary = 5
        self.X_train_Numeric=None
        self.X_test_Numeric=None
        self.X_train_Category=None
        self.X_train_Numeric=None
        self.y_train=None
        self.y_test=None
        super().__init__(dataframe, percentile)

    def ArquitectureModel(self):
        self.X_train_Numeric, self.X_test_Numeric, self.X_train_Category, self.X_test_Category,self.y_train,self.y_test=super().Split(y='Time_taken_min')

        vocab = list(self.X_train_Category)
        index_range = tf.range(len(vocab), dtype=tf.int64)
        table_init = tf.lookup.KeyValueTensorInitializer(vocab, index_range)
        table = tf.lookup.StaticVocabularyTable(table_init, self.out_of_vocabulary)

        InputNumeric = keras.layers.Input(shape=(6,))
        InputCategory = keras.layers.Input(shape=(2,), dtype=tf.string)


        cat_indices = keras.layers.Lambda(lambda cats: table.lookup(cats))(InputCategory)
        Embedings_dims = keras.layers.Embedding(input_dim=len(self.X_train_Category), output_dim=2)(cat_indices)
        flattened_embeddings = keras.layers.Flatten()(Embedings_dims)
        concatenated = keras.layers.concatenate([InputNumeric, flattened_embeddings], axis=-1)
        batch_norm=keras.layers.BatchNormalization()
        hidden1 = keras.layers.Dense(30, activation='relu', kernel_initializer='he_normal')(concatenated)
        batch_norm=keras.layers.BatchNormalization()(hidden1)
        hidden2 = keras.layers.Dense(30, activation='relu', kernel_initializer='he_normal')(batch_norm)
        hidden3 = keras.layers.Dense(30, activation='relu', kernel_initializer='he_normal')(hidden2)
        output = keras.layers.Dense(1)(hidden3)


        Model = keras.Model(inputs=[InputNumeric, InputCategory], outputs=[output])
        return Model

    def PlotArquitecture(self):
        Model = self.ArquitectureModel()
        return keras.utils.plot_model(Model, show_shapes=True, show_layer_names=True)

```

```python
class Transformer:
  def __init__(self,dataframe):
    self.dataframe=dataframe
    self.clean_data=CleanData(self.dataframe)
    self.preprocess=PreprocessData(self.dataframe)
    self.premodel=CreateArquitectureModel(self.dataframe,0.25)

  def transformer(self):
    self.dataframe=self.clean_data.IsInconsistence()
    self.dataframe=self.preprocess.Preprocess()


    return self.dataframe
```

```python
class TrainingModel(Transformer):
    def __init__(self, dataframe):
        super().__init__(dataframe)
        self.model = self.premodel.ArquitectureModel()
        self.model_history = None

    @tf.function
    def root_mean_squared_error(self, y_true, y_pred):
        return tf.sqrt(tf.reduce_mean(tf.square(tf.cast(y_true, dtype=tf.float32) - y_pred)))

    def Training(self):
      root_mean_squared_error=self.root_mean_squared_error
      self.model.compile(
            loss=self.root_mean_squared_error,
            optimizer=keras.optimizers.RMSprop(),
            metrics=[
                root_mean_squared_error
            ]
        )

      self.model_history = self.model.fit(
            (self.premodel.X_train_Numeric, self.premodel.X_train_Category),
            self.premodel.y_train,
            epochs=32,
            validation_data=((self.premodel.X_test_Numeric, self.premodel.X_test_Category), self.premodel.y_test),
        )

    def PlotErrorTestTraining(self):
        pd.DataFrame(self.model_history.history).plot(figsize=(6, 6))
        plt.grid(True)
```

```python
transformer=Transformer(dataframe)
```

```python
dataframe=transformer.transformer()
```
